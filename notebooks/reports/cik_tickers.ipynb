```python
DB_PATH = #CHECK in database_connector.py

connector = DatabaseConnector(DB_PATH)

filings = connector.fetch_data("filings")
scores = connector.fetch_data("scores")

with open("../financial_report_analyzer/defaults/missing_ticker_cik_mapping.yaml", "r") as f:
    cik_ticker_mapping = yaml.load(f, Loader=yaml.Loader)
```

```python
scraper = SECScraper()
```

```python
cik_filings = {}
for ticker, cik in cik_ticker_mapping.items():
    cik_filings[ticker] = scraper.get_10k_filings(cik)
```

```python
cik_filings = pd.DataFrame(cik_filings).T
```

```python
cik_filings
```

```python
years = cik_filings.columns.tolist()
```

```python
cik_filings = cik_filings[sorted(years)]
```

```python
filings
```

```python
cik_filings = cik_filings.melt(ignore_index=False).reset_index().rename(columns={"index":"ticker", "variable":"year", "value":"url"})
```

```python
def get_type(url):
    if not pd.isna(url):
        return url.split(".")[-1]
    return np.nan
```

```python
cik_filings["url_type"] = cik_filings["url"].apply(lambda x: get_type(x))
```

```python
cik_filings = cik_filings.sort_values(["ticker", "year"]).dropna().reset_index(drop=True)
```
