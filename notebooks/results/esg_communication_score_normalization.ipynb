# Prepare Data S&P 500 ESG

```python
sp500_esg_scores = pd.read_csv(r"/Users/timmuller/Documents/Uni/Master/Thesis/ESG_Scores/sp500esg_esg_scores_with_permno.csv").drop(columns="Unnamed: 0")
```

# Normalize ESG communication Scores of S&P 500 ESG

```python
def min_max_normalize(series):
    return (series - series.min()) / (series.max() - series.min())
```

```python
normalized_sp500_esg_scores = sp500_esg_scores.copy()
```

```python
columns_to_normalize = ['environmental', 'social', 'governance', 'esg']

# Apply normalization to each score
for column in columns_to_normalize:
    normalized_sp500_esg_scores[column + '_normalized'] = min_max_normalize(sp500_esg_scores[column])

normalized_sp500_esg_scores.drop(columns=columns_to_normalize, inplace=True)
normalized_sp500_esg_scores
```

```python
normalized_sp500_esg_scores.to_csv(r"/Users/timmuller/Documents/Uni/Master/Thesis/ESG_Scores/sp500esg_esg_scores_normalized_with_permno.csv")
```
