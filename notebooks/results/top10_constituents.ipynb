```python
import pandas as pd
```

# Define function to get top 10 Constituents

```python
def get_top_10_constituents(underlying_index: pd.DataFrame, company_info: pd.DataFrame) -> pd.DataFrame:
    
    if 'permno' in company_info.columns:
        identifier_to_name = dict(zip(company_info['permno'].astype(str), company_info['company name']))
    else:
        identifier_to_name = dict(zip(company_info['isin'].astype(str), company_info['company name']))

    # Rename the columns of underlying_index using the permno_to_name mapping
    underlying_index = underlying_index.rename(columns=identifier_to_name)
    index_overview = pd.DataFrame(index=underlying_index.index, columns=range(1, 11))

    for date, row in underlying_index.iterrows():
        # Get the top 10 values and their column names (permno)
        top_10 = row.nlargest(10)
        top_10_values = [f"{id} - {value * 100:.2f}%" for id, value in top_10.items()]
        index_overview.loc[date] = top_10_values
    
    return index_overview
```
