# Check for missing reports on rebalancing day

```python
sp_500_esg_index_composition['year'] = pd.to_datetime(sp_500_esg_index_composition['date']).dt.year
sp_500_esg_index_composition
```

```python
results = pd.DataFrame()
# Loop through each column (permno) in index_composition (ignoring the 'date' and 'year' columns)
for permno in sp_500_esg_index_composition.columns.drop(['date', 'year']):
    # Check if each permno has non-NaN values
    for index, row in sp_500_esg_index_composition.iterrows():
        if pd.notna(row[permno]):
            year = row['year']
            # Filter esg_scores for the current permno and year
            esg_sub = sp_500_esg_scores[(sp_500_esg_scores['permno'] == int(permno)) & (sp_500_esg_scores['year'] == int(year))]
            # Check if there are any ESG scores for the filtered subset
            if not esg_sub.empty:
                # If no scores exist, append the check but indicate no scores
                result = pd.DataFrame({
                    'permno': permno,
                    'date': row['date'],
                    'year checked': year,
                    'has scores': True
                }, index=[0])
                results= pd.concat([results, result], axis=0)
            else:
                result = pd.DataFrame({
                    'permno': permno,
                    'date': row['date'],
                    'year checked': year,
                    'has scores': False
                }, index=[0])
                results= pd.concat([results, result], axis=0)
```

```python
print(f"Number of needed annual reports over the reconstruction period: {sp_500_esg_index_composition.drop(columns=['date', 'year']).notna().sum().sum()}")
print(f"Number of companies that have annual report available when they are part of the index: {results['has scores'].sum()}")
print(f"Missing Reports: {sp_500_esg_index_composition.drop(columns=['date', 'year']).notna().sum().sum() - results['has scores'].sum()}")
```

# Check which companies have missing reports

```python
missing_anual_reports = results[results["has scores"]==False]
missing_anual_reports["permno"] = missing_anual_reports.loc[:, "permno"].astype(int)
df = sp_500_esg[["permno", "company name"]].drop_duplicates()
missing_anual_reports = missing_anual_reports.merge(df, on="permno", how="left")
missing_anual_reports
```

```python
dax_50_esg_index_composition['year'] = pd.to_datetime(dax_50_esg_index_composition['date']).dt.year
dax_50_esg_index_composition
```

```python
dax_50_esg_results = pd.DataFrame()
# Loop through each column (isin) in dax_index_composition (ignoring the 'date' and 'year' columns)
for isin in dax_50_esg_index_composition.columns.drop(['date', 'year']):
    # Check if each isin has non-NaN values
    for index, row in dax_50_esg_index_composition.iterrows():
        if pd.notna(row[isin]):
            year = row['year']
            # Filter esg_scores for the current isin and year
            dax_esg_sub = dax_50_esg_scores[(dax_50_esg_scores['isin'] == str(isin)) & (dax_50_esg_scores['year'] == int(year))]
            # Check if there are any ESG scores for the filtered subset
            if not dax_esg_sub.empty:
                # If no scores exist, append the check but indicate no scores
                dax_50_esg_result = pd.DataFrame({
                    'isin': isin,
                    'date': row['date'],
                    'year checked': year,
                    'has scores': True
                }, index=[0])
                dax_50_esg_results= pd.concat([dax_50_esg_results, dax_50_esg_result], axis=0)
            else:
                dax_50_esg_result = pd.DataFrame({
                    'isin': isin,
                    'date': row['date'],
                    'year checked': year,
                    'has scores': False
                }, index=[0])
                dax_50_esg_results= pd.concat([dax_50_esg_results, dax_50_esg_result], axis=0)
```

```python
print(f"Number of needed annual reports over the reconstruction period: {dax_50_esg_index_composition.drop(columns=['date', 'year']).notna().sum().sum()}")
print(f"Number of companies that have annual report available when they are part of the index: {dax_50_esg_results['has scores'].sum()}")
print(f"Missing Reports: {dax_50_esg_index_composition.drop(columns=['date', 'year']).notna().sum().sum() - dax_50_esg_results['has scores'].sum()}")
```
